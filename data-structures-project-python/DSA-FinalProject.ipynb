1. Import Libraries:

Imagine you're setting up a doctor's appointment system on your computer. This line adds tools you'll need to work with dates and times, like scheduling appointments for specific times.

```python
from datetime import datetime, timedelta
```

2. Function for scheduling appointments:

This function creates a schedule of appointments based on a specified start time and the duration of each appointment. It converts the start time into a datetime object, computes the end time for every appointment, and formats the time strings to follow the "%I:%M %p" format (hour:minute AM/PM). Finally, it returns the resulting schedule as a list of tuples.

```python
# Function to schedule appointments
def schedule_appointments(start_time, durations):
    start_datetime = datetime.strptime(start_time, "%I:%M %p")
    current_time = start_datetime
    schedule = []
    for duration in durations:
        end_time = current_time + timedelta(minutes=duration)
        schedule.append((current_time.strftime("%I:%M %p"), end_time.strftime("%I:%M %p")))
        current_time = end_time
    return schedule
```

3. Function for Waiting Time: 

This function computes the waiting time for each process within a round-robin scheduling algorithm. It iterates through the processes, continually adjusting the remaining burst time until all processes are finished. The waiting time for each process is saved in the "wt" list, while the order of executed processes is stored in the "order" list.

```python
def findWaitingTime(processes, n, bt, wt, quantum, order): 
    rem_bt = [0] * n
    for i in range(n): 
        rem_bt[i] = bt[i]
    t = 0 
    while(1):
        done = True
        for i in range(n):
            if (rem_bt[i] > 0) :
                done = False 
                if (rem_bt[i] > quantum) :
                    t += quantum 
                    rem_bt[i] -= quantum
                    order.append([processes[i], quantum])
                else:
                    order.append([processes[i], rem_bt[i]])
                    t = t + rem_bt[i] 
                    wt[i] = t - bt[i] 
                    rem_bt[i] = 0
        if (done == True):
            break
```

4. Function for turn around time:

This function calculates the turnaround time for each process in a scheduling algorithm. Turnaround time is defined as the total time taken by a process from its arrival in the system to its completion, including both waiting time and execution time. The function iterates through each process, calculates the turnaround time by adding the burst time (execution time) and the waiting time, and stores the result in the tat (turnaround time) array.



```python
# Function to calculate turn around time 
def findTurnAroundTime(processes, n, bt, wt, tat):
    for i in range(n):
        tat[i] = bt[i] + wt[i] 
 
```

5. Function for calculating Average Times:

This function findavgTime computes the average waiting time and turnaround time for a set of processes. It iterates through each process, calculates the waiting and turnaround times, and displays the process details including waiting and turnaround times. Finally, it prints the average waiting time, turnaround time, and the order of processes executed along with their appointment schedule.

```python
# Function to calculate average waiting 
# and turn-around times. 
def findavgTime(processes, n, bt, quantum):  
    wt = [0] * n
    tat = [0] * n 
    order = [] # to store the order of processes getting executed

    findWaitingTime(processes, n, bt, wt, quantum, order) 
    findTurnAroundTime(processes, n, bt, wt, tat) 

    # Display processes along with all details 
    print("\nPatients\tApproc VisitTime\tWaitingTime\tTurn-Around Time")
    total_wt = 0
    total_tat = 0
    for i in range(n):
        total_wt = total_wt + wt[i] 
        total_tat = total_tat + tat[i] 
        print(" ", processes[i][0], "\t\t", bt[i], 
              "\t\t", wt[i], "\t\t", tat[i])
 
    print("\nAverage waiting time = %.5f "%(total_wt /n) )
    print("Average turn around time = %.5f "% (total_tat / n))
    print("\n\nOrder of processes executed:")
    for p in order:
        print(f" {p[0][0]}: {p[1]} mins")
     
    start_time = "8:00 AM"
    appointments = schedule_appointments(start_time, [p[1] for p in order])

    # Print the schedule
    print("\nSchedule:")
    # Print the schedule with patient names
    for i, (name, (start, end)) in enumerate(zip(order, appointments), start=1):
        print(f"Patient {i}: {name[0][0]} - {start} to {end}")
```

5. Priority Scheduling:

This function helps manage patient appointments based on priority. It arranges patients in order, taking into account their priority level and when they arrived. Then, it goes through the list, showing the order in which patients will see the doctor. Additionally, it calculates and shares the average waiting time and turnaround time for the scheduled appointments.

```python
def priorityScheduling(proc, n):
      
    approx_time = []
    final_list = []
    # Sort processes by priority first,then approx time
    proc = sorted(proc, key=lambda proc: (proc[2],proc[1]),reverse=True ) 
    quantum = 60

    print(" \n \n Order in which patients gets attended by doctor") 
    for i in proc: 
        print( i[0] + ',', end=" ")
        final_list.append(i[0])
        approx_time.append(i[1])
    
    findavgTime(proc, n,approx_time,quantum)
```

6. Driver Code

This part asks individuals to provide details about their visit, such as their name and reason for coming. After determining the purpose, it estimates the time needed for the visit and assigns a priority level. Patients are then included in a list, and if there's room during the working hours, they're placed in line to see the doctor. Once the queue is full or the day's maximum working hours are reached, the scheduling function is invoked to prioritize patient visits and carry out the scheduling process.

```python
# Driver code     
patientslist = []
doctorsqueue = []
time = 0
approxtime = 0
priority = 0
while True:
    name = input('Enter patient Name: ')
    purpose = input('Pick visit purpose: \n 1)Normal-(Healthcheckup)\n 2)Emergency(Operations-Accidents,Fractures)\n 3)Issues(Fever,flu,cough,headache,diahhera)\n')
    
    if purpose == '1':
        approxtime = 30
        priority = 0
    elif purpose == '2':
        approxtime = 90
        priority = 2
    elif purpose == '3':
        approxtime = 45
        priority = 1
    else:
        print('Wrong Input')

    patientslist.append([name, approxtime, priority])    
    if time < 480 and (time + approxtime) < 480:
        doctorsqueue.append([name, approxtime, priority])
        time = time + approxtime
    else:
        print("\nReached max number of patients today. Please try tomorrow.", name)
        break
        
n = len(doctorsqueue)
priorityScheduling(doctorsqueue, n)
```
